Import part :

```python
# the four different states of the XOR gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```python
# first layer in the sequential model
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
# Now we know that the model will take as input array of two dimension
# and output arrays of shape (None, 16).
# This will be our first line of hidden neutrons which follow the rule of activation relu
# (more details about it in our lessons)
model.add(Dense(1, activation='sigmoid'))
# And the final result will be in a array of shape (None, 1)
```

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

model.fit(training_data, target_data, epochs=500, verbose=2) 

print (model.predict(training_data).round())
```

Extension of this into four different gates :

The AND gate :

```python
# the four different states of the AND gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[0],[0],[0],[1]], "float32")

modelAnd = Sequential()
modelAnd.add(Dense(16, input_dim=2, activation='relu'))
modelAnd.add(Dense(1, activation='sigmoid'))

modelAnd.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

modelAnd.fit(training_data, target_data, epochs=500, verbose=2) 

print (modelAnd.predict(training_data).round())
```

The OR gate :

```python
# the four different states of the OR gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[0],[1],[1],[1]], "float32")

modelOR = Sequential()
modelOR.add(Dense(16, input_dim=2, activation='relu'))
modelOR.add(Dense(1, activation='sigmoid'))

modelOR.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

modelOR.fit(training_data, target_data, epochs=500, verbose=2) 

print (modelOR.predict(training_data).round())
```

The NOT gate

```python
# the two different states of the Not gate
training_data = np.array([[0],[1]], "float32")

# the two expected results in the same order
target_data = np.array([[1],[0]], "float32")

modelNot = Sequential()
modelNot.add(Dense(16, input_dim=1, activation='relu'))
modelNot.add(Dense(1, activation='sigmoid'))

modelNot.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

modelNot.fit(training_data, target_data, epochs=500, verbose=2) 

print (modelNot.predict(training_data).round())
```

The NAND gate :

```python
# the four different states of the AND gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[1],[1],[1],[0]], "float32")

modelNAnd = Sequential()
modelNAnd.add(Dense(16, input_dim=2, activation='relu'))
modelNAnd.add(Dense(1, activation='sigmoid'))

modelNAnd.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

modelNAnd.fit(training_data, target_data, epochs=500, verbose=2) 

print (modelNAnd.predict(training_data).round())
```

The epoch value in the fonction fit correspond to the number of time you go through youre training set.
Epochs affect the training, because we can see that the accuracy value increase at each iteration. But when we observe all the results we discover that the accuracy value seem to stagnate after a certain point, so increasing enven more the value of the epoch is meaningless.
